```python
from double_pendulum.model.symbolic_plant import SymbolicDoublePendulum
from double_pendulum.simulation.simulation import Simulator
from double_pendulum.controller.pid.point_pid_controller import PointPIDController
from double_pendulum.utils.plotting import plot_timeseries
import numpy as np

plant = SymbolicDoublePendulum(mass=[0.6, 0.5], length=[0.3, 0.2])

sim = Simulator(plant=plant)
controller = PointPIDController(torque_limit=[5.0, 5.0])
controller.set_parameters(Kp=50, Ki=1, Kd=10)
controller.init()

# SIMULA SENZA ANIMATE (veloce+stabile)
T, X, U = sim.simulate(t0=0.0, x0=[2.8, 1.2, 0., 0.], tf=10.0, dt=0.01, controller=controller)

T_arr = np.array(T)
X_arr = np.array(X)
U_arr = np.array(U)

print(f"SIM OK: T={T_arr.shape}, U={U_arr.shape}")
plot_timeseries(T_arr, X_arr, U_arr, show=True, save_to="swingup.png")
```
