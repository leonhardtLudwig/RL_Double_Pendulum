```python
from double_pendulum.controller.pid.point_pid_controller import PointPIDController
import numpy as np

controller = PointPIDController(torque_limit=[5.0, 5.0])
controller.set_parameters(Kp=50, Ki=0, Kd=10)
controller.init()
x_test = [2.8, 1.2, 0., 0.]
u = controller.get_control_output(x_test, 0.0)
print(f"TEST CONTROLLER: u = {u}")  # Deve ~[5.0, -5.0]

```

```python
from double_pendulum.model.symbolic_plant import SymbolicDoublePendulum

plant = SymbolicDoublePendulum(mass=[0.6, 0.5], length=[0.3, 0.2])
x_test = [2.8, 1.2, 0., 0.]
rhs = plant.rhs(0.0, x_test, [0., 0.])
print(f"TEST PLANT RHS: finite={np.all(np.isfinite(rhs))}, valore={rhs}")
```

```python
from double_pendulum.simulation.simulation import Simulator

sim = Simulator(plant=plant)
controller = PointPIDController(torque_limit=[5.0, 5.0])
controller.set_parameters(Kp=50, Ki=0, Kd=10)
controller.init()

T, X, U = sim.simulate(t0=0.0, x0=[2.8, 1.2, 0., 0.], tf=2.0, dt=0.01, controller=controller)
print(f"TEST SIM: T steps={len(T)}, U shape={np.array(U).shape}")
```
